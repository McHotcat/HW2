```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

## Problem 3.1

```python
X1 = np.array([1,-2,-1,-1,1])
X2 = np.array([2,-1,2,0,-2])
X3 = np.array([-1,0,2,2,1])
y1 = -7
y2 = -1
y3 = -1
```

```python
X = np.array([X1,X2,X3])
X
```

```python
y = np.array([y1,y2,y3]).reshape([3,1])
y
```

```python
weight = np.array([[0,0,0,0,0]]).T
weight
```

```python
w0 = []
w1 = []
w2 = []
w3 = []
w4 = []
W = [w0,w1,w2,w3,w4]
L = []
```

```python
for epoch in range(200):
    loss = np.ones(3).dot((X.dot(weight) - y)**2)
    grad = X.T.dot(X.dot(weight)-y)
    weight = weight - 0.02*grad
    L.append(np.log(loss[0]))
    for i in range(5):
        W[i].append(weight[i][0])
```

```python
plt.plot(range(200),L)
```

```python
plt.plot(range(200),w0,color='r')
plt.plot(range(200),w1,color='g')
plt.plot(range(200),w2,color='b')
plt.plot(range(200),w3,color='y')
plt.plot(range(200),w4,color='c')
```

## Problem 3.2 

```python
w0 = []
w1 = []
w2 = []
w3 = []
w4 = []
W = [w0,w1,w2,w3,w4]
L = []
```

```python
weight = np.array([[0,0,0,0,0]]).T
weight
```

```python
for epoch in range(200):
    loss = np.ones(3).dot((X.dot(weight) - y)**2)
    grad = X.T.dot(X.dot(weight)-y)
    weight = weight - 0.02*grad
    weight_1d = weight.flatten()
    sorted_weight = np.sort(abs(weight_1d))
    threshold = sorted_weight[3]
    for i in range(5):
        if abs(weight_1d[i]) < threshold:
            weight_1d[i] = 0
    weight = weight_1d.reshape([5,1])
    L.append(np.log(loss[0]))
    for i in range(5):
        W[i].append(weight[i][0])
```

```python
plt.plot(range(200),L)
```

```python
plt.plot(range(200),w0,color='r')
plt.plot(range(200),w1,color='g')
plt.plot(range(200),w2,color='b')
plt.plot(range(200),w3,color='y')
plt.plot(range(200),w4,color='c')
```

## Problem 3.3 

```python
w0 = []
w1 = []
w2 = []
w3 = []
w4 = []
W = [w0,w1,w2,w3,w4]
L = []
```

```python
weight = np.array([[0,0,0,0,0]]).T
weight
```
